```python
from datasets import load_dataset
from transformers import pipeline
```

```python
# Load a pre-trained sentiment analysis model
sentiment_analyzer = pipeline("sentiment-analysis")
```

```python
# Load Amazon reviews dataset
# Using Hugging Face datasets library
dataset = load_dataset("mteb/amazon_polarity")
```

```python
# Select a subset of reviews for analysis
# Select a subset of reviews for analysis
reviews = dataset['test']['text'][:10]
```

```python
# Perform sentiment analysis
results = sentiment_analyzer(reviews)
```

```python
print(results)
```

```python
# Display the results
for i, review in enumerate(reviews):
    print(f"Review {i+1}: {review}")
    print(f"Sentiment: {results[i]['label']} (Score: {results[i]['score']:.2f})")
    print("-" * 80)
```
